# About the dataset :-

The dataset is available at Kaggle : https://www.kaggle.com/datasets/yasserh/titanic-dataset

The sinking of the Titanic is one of the most infamous shipwrecks in history.

On April 15, 1912, during her maiden voyage, the widely considered “unsinkable” RMS Titanic sank after colliding with an iceberg. Unfortunately, there weren’t enough lifeboats for everyone on board, resulting in the death of 1502 out of 2224 passengers and crew.

While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others.

In this challenge,I built a predictive model that answers the question: “what sorts of people were more likely to survive?” using passenger data (ie name, age, gender, socio-economic class, etc).

```python
## Importing Libaries to be used
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

#  Importing Data To Be Used

```python
data=pd.read_csv("D://All Datas//csv files//Titanic-Dataset.csv")
data
```

```python
data.describe()
```

# Exploring Missing Data 

```python
data.isnull().sum()
```

```python
data.drop(["Age","Cabin"],axis=True,inplace=True)
data
```

```python
data.dropna(inplace=True)
data
```

```python
data.isnull().sum()
```

#  Data Visualization

```python
sns.set()
```

# Label Encoding

```python
data.drop(["Name","Ticket"],axis=True,inplace=True)
data
```

```python
l1=LabelEncoder()
data['Sex']= l1.fit_transform(data['Sex'])
data['Embarked']= l1.fit_transform(data['Embarked'])
```

```python
data
```

# Data Splitting

```python
y=data["Survived"]
x=data.drop("Survived",axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
y_train
```

# Model Selection

```python
from sklearn.metrics import accuracy_score
```

```python
models=[]
algorithem=[]
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
acc=accuracy_score(y_test, y_pred)
acc
```

```python
models.append(acc)
print(models)
algorithem.append("knn")
print(algorithem)
```

# Decision Tree

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = tree.DecisionTreeClassifier(max_depth=10)
y_pred = dt.fit(x_train, y_train).predict(x_test)
acc2=accuracy_score(y_test, y_pred)
acc2
```

```python
models.append(acc2)
print(models)
algorithem.append("Decision Tree")
print(algorithem)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier(max_depth=2, random_state=0)
y_pred = rf.fit(x_train, y_train).predict(x_test)
acc3=accuracy_score(y_test, y_pred)
acc3
```

```python
rf.feature_importances_
```

```python
x_train.columns
```

```python
pd.DataFrame((rf.feature_importances_*100),x_train.columns)
```

```python
models.append(acc3)
print(models)
algorithem.append("Random Forest")
print(algorithem)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
clf=SVC()
y_pred = clf.fit(x_train, y_train).predict(x_test)
acc4=accuracy_score(y_test, y_pred)
acc4
```

```python
models.append(acc4)
print(models)
algorithem.append("SVM")
print(algorithem)
```

```python
plt.bar(algorithem,models)
plt.show()
```
